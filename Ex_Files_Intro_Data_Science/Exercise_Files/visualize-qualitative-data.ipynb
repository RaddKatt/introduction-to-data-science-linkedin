```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv('Airbnb_NYC_2019.csv')
```

```python
listings
```

```python
"""
Use the countplot method to create the bar chart
using the 'listings' DataFrame. Setting x = 'neighborhood_group'
indicates that we want to use data from the neighborhood_group column.

Use matplot
"""
sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
# Create a bar chart
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings)

# Show the bar chart that is created
plt.show()

# Note: the black lines are confidence intervals
# that were generated by the barplot method.
```

```python
# Recreate the above Bar Chart without
# the black lines (confidence intervals)
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```
