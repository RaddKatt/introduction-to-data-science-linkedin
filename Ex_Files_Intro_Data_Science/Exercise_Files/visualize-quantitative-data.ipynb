## Quantative Data Visualization
**Quantitative data** is another term for *numerical data*.<br />
**Numerical data** is data that consists of *amounts* or *quantities*.

Two types of charts that are commonly used to visualize **quantitative** data are **Histograms** and **Scatter Plots**.


### Import Python Libraries

```python
import pandas as pd              # Library for Python data analysis
import matplotlib.pyplot as plt  # Library to create figures and plots
import numpy as np               # Library for scientific computing
```

### Create a Pandas DataFrame

```python
# Read in the csv file and store the pandas DataFrame
# in the variable named 'listings'
listings = pd.read_csv('Airbnb_NYC_2019.csv')
```

```python
# View the Airbnb listings dataset
listings
```

### Create a Simple Histogram from the Pandas DataFrame

```python
plt.hist(listings['price'])         # Create a histogram
plt.xlabel('Price (in US Dollars)') # Create a label for the X-axis
plt.show()                          # Show the histogram
```

### Create a Better Histogram from the Pandas DataFrame
Use bins to make the visualization easier.<br />
* 0,1100 is the x range, and  40 is the bin width.

```python
plt.hist(listings['price'], bins = np.arange(0, 1100, 40))
plt.xlabel('Price (in US Dollars)')
plt.show()
```

### Create a Scatter Plot from the Pandas DataFrame
Scatter Plots are often used to compare two sets of quantitative data.

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.show()
```

### Restrict the x-axis of the Scatter Plot to Prices of $1100
* Use the *matplotlib.pyplot* **xlim** method

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.xlim(0, 1100)     # Restrict the x-axis to a price of 1100
plt.show()
```

### Decrease the Size of the Scatter Plot's Points
* Use the `s` parameter of the *matplotlib.pyplot* **scatter** method.
* `plt.scatter(x = X['price'], y = X['number_of_reviews'], s = 5)`

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s = 5) # Decrease the size of the scatter plot's points
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.xlim(0, 1100)
plt.show()
```

It appears that listings with lower prices have more reviews.<br />
This leads to a question: "Is there an association between the price and number of reviews for listings"?

Visualizing your data helps you to observe trends.
