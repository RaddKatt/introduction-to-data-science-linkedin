```python
# import pandas library

import pandas as pd
```

```python
us_babies = pd.read_csv('us_baby_names.csv')
```

```python
us_babies
```

```python
def popular(s):
    """Receives s, a Pandas Series, containing baby names in order of highest count to lowerst count.
    Returns the most popular name in s."""
    return s.iloc[0]
```

```python
us_babies_grouped = us_babies.sort_values('Count', ascending = False).groupby(['Year','Gender']).agg(popular)
```

```python
us_babies_grouped
```

```python
us_michael = us_babies.loc[us_babies['Name'] == 'Michael', :]
```
